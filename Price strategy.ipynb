```python
# Fit a random forest classifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(df.drop('price_range', axis=1), df['price_range'])

# Get feature importances
feature_importances = rf.feature_importances_

# Plot feature importances
plt.barh(df.drop('price_range', axis=1).columns, feature_importances)
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importance for Predicting Price Range')
plt.show()
```

```python
#splitting data into training and testing
X=df.drop('price_range',axis=1)
y=df['price_range']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

# First model

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
a=lm.score(X_test,y_test)
print(a)
```

# Second model

```python
knn = KNeighborsRegressor(n_neighbors=10)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
b=knn.score(X_test, y_test)
print(b)
```

# Third model

```python
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train, y_train)
c=rfc.score(X_test,y_test)
print(c)
```

# Fourth model

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
d=logmodel.score(X_test,y_test)
print(d)
```

# Test Data

```python
df1=pd.read_csv("test.csv")
df1.head()
df1 = df1.drop('id', axis=1)
```

```python
X=df.drop('price_range',axis=1)
y=df['price_range']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
# here we split the test data set so we can validate our results.
```

```python
# Check if the 'predicted_price_range' column exists before trying to drop it
if 'predicted_price_range' in df1.columns:
    df1 = df1.drop('predicted_price_range', axis=1)
```

```python
# Since KNN provides the highest R2 score, we select that model to run on the test dataset
knn = KNeighborsRegressor(n_neighbors=10)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
knn.score(X_test, y_test)
```

```python
# Display only the predictions
print(df1['predicted_price_range'])
```

```python
print(df1['predicted_price_range'].describe())
```

```python
# Assuming df1 is your testing dataset and it is ready for prediction
y_test_pred = knn.predict(df1)

# Round the predicted values
y_test_pred_int = np.round(y_test_pred).astype(int)

# If you want to add these predictions back to df1 to see them next to the original data:
df1['predicted_price_range'] = y_test_pred_int
```

```python
sns.histplot(y_test_pred_int, kde=True)
plt.title('Distribution of Predicted Price Ranges')
plt.xlabel('Predicted Price Range')
plt.ylabel('Frequency')
plt.show()
```
